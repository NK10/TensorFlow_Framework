# Boxplot

This will show the boxplot not nicely becuase of the variation in the data values for each column, so we can use the 
the normalization or standardization to bring all variables on the same scale.

•When the box plot is short it implies that much of your data points are similar, since there are many values in a small range

•When the box plot is tall it implies that much of your data points are quite different, since the values are spread over a wide range

•If the median value is closer to the bottom then we know that most of the data has lower values. If the median value is closer to the top then we know that most of the data has higher values. Basically, if the median line is not in the middle of the box then it is an indication of skewed data.

•Are the whiskers very long? That means your data has a high standard deviation and variance i.e the values are spread out and highly varying. If you have long whiskers on one side of the box but not the other, then your data may be highly varying only in one direction.

```python
def missing_count(df,threshold= 0):
    """
    #input: 
        df = dataframe for which we have to calculate the missing counts
        threshold = threshold for the missing count
    #return:
        columns and their missing count percentage"""
    missing  = df.isnull().sum().sort_values(ascending = False)
    return (missing[missing>len(df)*(threshold/100)]/len(df))*100
```

```python
def fill_missing(df, logic = "mean",threshold = 0):
    df_missing = missing_count(df,threshold)
    #print(df_missing)
    for i in range(len(df_missing)):
        
        if(df[df_missing.index[i]].dtypes== "int64" or df[df_missing.index[i]].dtypes== "float64"):
            #print("int",df_missing.index[i])
            df[df_missing.index[i]].fillna(np.mean(df[df_missing.index[i]]), inplace = True)
        elif(df[df_missing.index[i]].dtypes== "object"):
            df[df_missing.index[i]].fillna(df[df_missing.index[i]].mode()[0], inplace = True)
        else:
            print("other",df_missing.index[i])
    return df
# DataFrame.interpolate need to explore 
    
    
```

```python
def outlier_count(df,col_name):
    """
    #input: 
        df = dataframe for which we have to calculate the missing counts
        col_name = name of the column in the dataframe for which we have to calculate the IQR
    #return:
        Count of records"""
    Q75 , Q25 = df[col_name].quantile([.75,.25])
    iqr = Q75 - Q25
    mini = Q25 - (iqr*1.5)
    maxm = Q75 + (iqr*1.5)
    return len(df[(df[col_name] <mini) | (df[col_name] > maxm)])
```

```python
def outlier_count_df(df):
    outlier_counts = []
    int_con, float_con , _, _=  dtypes_columns(df)
    for i in range(len(int_con)):
        outlier_cnt = []

        #col_na = str(int_con[i])
        count = outlier_count(df,int_con[i])
        outlier_cnt.append(int_con[i])
        outlier_cnt.append(count)
        outlier_counts.append(outlier_cnt)
    for i in range(len(float_con)):
        outlier_cnt = [] 
        #col_na = str(int_con[i])
        count = outlier_count(df,float_con[i])
        outlier_cnt.append(float_con[i])
        outlier_cnt.append(count)
        outlier_counts.append(outlier_cnt)
    return outlier_counts
```

```python
def label_counts(df):
    _,_,cat,_ = dtypes_columns(df)
    cat_counts = []
    for i in range(len(cat)):
        for j in range(len(df[cat[i]].value_counts())):
            cat_count = []
            cat_count.append(cat[i])
            cat_count.append(df[cat[i]].value_counts().index[j])
            cat_count.append(df[cat[i]].value_counts().values[j])
            cat_counts.append(cat_count)
    return cat_counts
```

```python
def cross_val_score_cv(model, X,Y,scor, cvv=5,njob= -1, withparameters = False):
    """ input 
        model = list of all the models for which we have to do the cross val
        X= independent variables dataframe
        Y = dependent variable series
        cvv = cross validation iteration
        njobs = parallel processing
        withparameters = if True will return the models parameters along with their score else just the name """

    from sklearn.model_selection import cross_val_score
    scores = []
    for name in model:
        score = []
        if withparameters : 
            model_name  = name
        else:
            model_name = type(name)
        score.append(model_name)
        val = cross_val_score(name, X,Y,scoring= scor , cv= cvv, n_jobs= njob)
        score.append(val)
        score.append("Mean score is "+ str(np.mean(val)))
        scores.append(score)

    return scores
```
