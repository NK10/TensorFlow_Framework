```python
iris = sns.load_dataset("iris")
```

```python
g = sns.pairplot(iris)
```

## Show different levels of a categorical variable by the color of plot elements:

```python
g = sns.pairplot(iris, hue="species")
```

## Use different markers for each level of the hue variable:

```python
g = sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
```

## Plot a subset of variables:

```python
g = sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
```

## Draw larger plots:2

```python
g = sns.pairplot(iris, height=3,
...                  vars=["sepal_width", "sepal_length"])
```

## Plot different variables in the rows and columns:

```python
g = sns.pairplot(iris,
...                  x_vars=["sepal_width", "sepal_length"],
...                  y_vars=["petal_width", "petal_length"])
```

## Plotting Multiple Graphs on a Grid

```python
g = sns.FacetGrid(tips, col="time", row="smoker")

g = g.map(plt.scatter, "total_bill", "tip", edgecolor="w")
```

```python
import matplotlib as plt
from scipy import stats
```

```python
#The top level container for all the plot elements.
fig = plt.pyplot.figure(figsize=(21,15)) # its width and height in the figsize 
#Specifies the geometry of the grid that a subplot can be placed in.
grid = plt.gridspec.GridSpec(3,4,figure=fig)
#location on the grid where the plot will visible
ax1 = fig.add_subplot(grid[0, 0:2])
ax1.set_title('boxplot')
sns.boxplot(tips.total_bill,)
ax2 = fig.add_subplot(grid[1:, 0:2])
ax2.set_title('Histogram/DistributionPlot')
sns.distplot(tips.total_bill,)
ax3 = fig.add_subplot(grid[0:33,2:4])
t =stats.probplot(tips.total_bill, plot=sns.mpl.pyplot)
```
