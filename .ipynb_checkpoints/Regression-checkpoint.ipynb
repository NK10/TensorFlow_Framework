```python
def missing_count(df,threshold):
    """ This function takes two arguments:
    input
        df = Dataframe
        threshold= fraction of column data in percent( between 0 and 100) having the missing values
    output
    data frame having the column name as index and missing % as the fraction missing the data"""
    if (isinstance(df, pd.DataFrame)):
        missing_data = pd.DataFrame( (df.isnull().sum().sort_values(ascending= False)/len(df))*100, columns=["Missing %"])
        return (missing_data[missing_data["Missing %"]> threshold])
    else:
        print("Passed data set is not a Dataframe!")
        
```

```python
def datatypes(df):
    '''This function returns the counts of various datatype in a dataframe
    input
        df = dataframe for which to count the datatypes
    output
        obj = datatype of Object type
        nonobj = datatype of Int or Float type
        unknown = datatype other than Object or Int or Float
    '''
    obj = []
    nonobj = []
    unknown = []
    if(isinstance(df, pd.DataFrame)):
        tpe = df.dtypes
        for i in range(len(tpe)):
            if tpe[i] =="object":
                obj.append(tpe.index[i])
            elif (tpe[i] =="int64" or tpe[i] =="float64"):
                nonobj.append(tpe.index[i])
            else:
                unknown.append(tpe.index[i])
        print("Total Object Type is {}\nTotal Float or Int is {}\nTotal unknown type is {}".format(len(obj),len(nonobj),len(unknown)))
        return obj, nonobj,unknown
    else:
        print("Passed dataframe is not a Dataframe")
        return obj, nonobj, unknown
```

```python
def skew_kurtosis(df):
    '''This function returns the kurtosis, skewness and std of a dataframe
    input
        df = dataframe for which to count the datatypes
    output
        dataframe having three columns as skewness, kurtosis and std along with the column name.
    '''
    if(isinstance(df, pd.DataFrame)):
        _, nobj,_  = datatypes(df)
        dictionary ={"Skewness":df[nobj].skew(), "Kurtosis":df[nobj].kurt(), "Std":df[nobj].std()}
        return (pd.DataFrame(dictionary))

    else:
        print("Passed dataframe is not a Dataframe")
```

```python
def fill_missing(df,threshold):
    '''This function returns the dataframe with no missing values
    input
        df = dataframe for which to fill the missing values
    output
        dataframe with filled missing values.
    '''
    if(isinstance(df,pd.DataFrame)):
        if(threshold >=0 and threshold <= 100):
            col = missing_count(df,threshold).index
            for i in col:
                if ((df[i].dtypes == "float64") or (df[i].dtypes == "int64") ):
                    df[i].fillna(df[i].mean(), inplace = True)
                else:
        # always be cautious with Mode, as The problem here is that mode returns a series and 
        #this is causing the fillna to faill, as mode returns a series so you have to access 
        #the first element as shown below with [0]
                    df[i].fillna(df[i].mode()[0], inplace = True)
                    #print(df[i].unique())
            return df
        else:
            print("Threshold should be in between 0-100")
    else:
        print("Passed dataframe is not a Dataframe")
```

```python
def unique_values(df):
    '''This function returns the unique values in a dataframe
    input
        df = dataframe for which to find the unique values
    output
        dataframe with the details of each object column, total_unique_count for that column
        , total_unique_values_for_that_column, unique_value, unique_value_record_counts.
    '''
    if(isinstance(df, pd.DataFrame)):
        obj,_,_ = datatypes(df)
        items = []
        for i in obj:
            g = df.groupby(df[i])
            for ke,val in g.groups.items():
                values =[]
                values.append(i)
                values.append(len(g.groups.keys()))
                values.append(g.groups.keys())
                values.append(ke)
                values.append(len(val))
                items.append(values)
            q = pd.DataFrame(items,columns=["col_name","col_unq_val_count",
                                            "col_total_unq_val",
                                            "col_unq_val",
                                            "unq_val_count"])
        return(q)
    else:
        print("Passed dataframe is not a Dataframe")
```

```python
from sklearn.linear_model import (LinearRegression, Lasso,Ridge,SGDRegressor
                                  ,ElasticNet)
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import GridSearchCV, cross_val_score , KFold
import matplotlib as plt
import seaborn as sns
```

```python
def model_output(mod, tr_x,te_x,tr_y,te_y):
    mod.fit(tr_x,tr_y)
    pred_test = mod.predict(te_x)
    pred_train = mod.predict(tr_x)
    error_test = mean_squared_error(te_y,pred_test)
    error_train = mean_squared_error(tr_y,pred_train)
    r_squared_train = mod.score(tr_x,tr_y)
    r_squared_test = mod.score(te_x,te_y)
    adj_r_sqr_train = 1 - (1-r_squared_train)*(len(tr_y)-1)/(len(tr_y)-tr_x.shape[1]-1)
    adj_r_sqr_test = 1 - (1-r_squared_test)*(len(te_y)-1)/(len(te_y)-tr_x.shape[1]-1)
    print("Train score is: {}\ntest score is: {}".format(r_squared_train,r_squared_test))
    print("adj r2 train is : {}\nadj r2 test is : {} ".format(adj_r_sqr_train,adj_r_sqr_test ))
    print("Train error is:{}\ntest error is: {}".format(error_train,error_test))
    
    
    #The top level container for all the plot elements.
    fig = plt.pyplot.figure(figsize=(8,10)) # its width and height in the figsize 
    #Specifies the geometry of the grid that a subplot can be placed in.
    grid = plt.gridspec.GridSpec(2,1,figure=fig)
    #location on the grid where the plot will visible
    ax1 = fig.add_subplot(grid[0, 0])
    ax1.set_title('Train Residual Plot')
    sns.residplot(tr_y, pred_train)
    ax2 = fig.add_subplot(grid[1:, 0])
    ax2.set_title('Test Residual Plot')
    sns.residplot(te_y,pred_test)
```

```python
mod = LinearRegression(normalize=True)
model_output(mod, train_x,test_x,train_y,test_y)
```
