# Retails Sales Data Analysis

```python
retaildata = pd.read_csv("C:\\Users\\USX28939\\Downloads\\onlineretail\\OnlineRetail.csv",encoding = 'unicode_escape')
```

```python
retaildata.head(5)
```

```python
retaildata.shape
```

```python
retaildata["Date"] = pd.to_datetime(retaildata["InvoiceDate"], format = '%m/%d/%Y %H:%M')
retaildata["Year"] = pd.DatetimeIndex(retaildata["Date"]).year
retaildata["Month"] = pd.DatetimeIndex(retaildata["Date"]).month
retaildata["Qtr"] = retaildata["Date"].dt.quarter
retaildata["Year_Qtr"] = retaildata["Date"].map(lambda x: 100*x.year+x.month)
retaildata["Sale"] = retaildata["Quantity"]*retaildata["UnitPrice"]
```

```python
retaildata.head()
```

```python
retaildata.groupby("Country").count()
```

```python
retaildata.groupby(['Year_Qtr'])['Sale'].count()
```

# Monthly Revenue ( NorthStar Metrics)

```python
tx_revenue = retaildata.groupby(['Year_Qtr'])['Sale'].sum().reset_index()
tx_revenue
```

```python
plt.figure(figsize=(12,6))
sb.barplot(x=tx_revenue["Year_Qtr"], y=tx_revenue["Sale"] )
```

```python
plt.figure(figsize=(12,6))
sb.pointplot(x=tx_revenue["Year_Qtr"], y=tx_revenue["Sale"] )
```

# Function Pct_change
DataFrame.pct_change(self: ~FrameOrSeries, periods=1, fill_method='pad', limit=None, freq=None, **kwargs)   
Percentage change between the current and a prior element.

```python
tx_revenue['MonthlyGrowth'] = tx_revenue['Sale'].pct_change()
```

```python
# is needed for the pointplot to work
tx_revenue.fillna(0, inplace=True)
```

```python
tx_revenue.head()
```

```python
plt.figure(figsize=(12,6))
sb.pointplot(x=tx_revenue["Year_Qtr"], y=tx_revenue["MonthlyGrowth"],join = True )
plt.title('Line Growth of Monthy Chart')
# Set x-axis label
plt.xlabel('Monthly Growth')
# Set y-axis label
plt.ylabel('Pct Change')
```

# New Customer Ratio (Other Metrics)

A good indicator of if we are losing our existing customers or unable to attract new ones

First we should define what is a new customer. In our dataset, we can assume a new customer is whoever did his/her first purchase in the time window we defined. We will do it monthly for this example.

```python
tx_uk.query("CustomerID == '17850' & Quantity > 7")
```

```python
tx_uk.groupby(["CustomerID"])["Date"].min()
```

```python
#create a dataframe contaning CustomerID and first purchase date
tx_min_purchase = tx_uk.groupby('CustomerID').Date.min().reset_index()
tx_min_purchase.columns = ['CustomerID','MinPurchaseDate']
tx_min_purchase['MinPurchaseYearMonth'] = tx_min_purchase['MinPurchaseDate'].map(lambda date: 100*date.year + date.month)

#merge first purchase date column to our main dataframe (tx_uk)
tx_uk = pd.merge(tx_uk, tx_min_purchase, on='CustomerID')

tx_uk.head()
```

```python
#create a column called User Type and assign Existing 
#if User's First Purchase Year Month before the selected Invoice Year Month
tx_uk['UserType'] = 'New'
tx_uk.loc[tx_uk['Year_Qtr']>tx_uk['MinPurchaseYearMonth'],'UserType'] = 'Existing'
tx_uk.head(400)
```

```python
#calculate the Revenue per month for each user type
tx_user_type_revenue = tx_uk.groupby(['Year_Qtr','UserType'])['Sale'].sum().reset_index()
tx_user_type_revenue.drop([0], inplace=True)
tx_user_type_revenue.head()
```

```python
temp = tx_uk.groupby(['Year_Qtr','UserType'], as_index=False,)['Sale'].sum()
temp.head()
```

```python
plt.figure(figsize = (10,6))
sb.pointplot(x=tx_user_type_revenue["Year_Qtr"], y =tx_user_type_revenue["Sale"] , 
             hue=tx_user_type_revenue["UserType"], join = True)
plt.title("New vs Existing")
plt.xlabel("Month")
plt.ylabel("Revenue")
```

# New Customer Ratio

```python
#create a dataframe that shows new user ratio - we also need to drop NA values (first month new user ratio is 0)
tx_user_ratio = tx_uk.query("UserType == 'New'").groupby(['Year_Qtr'])['CustomerID'].nunique()/tx_uk.query("UserType == 'Existing'").groupby(['Year_Qtr'])['CustomerID'].nunique() 
tx_user_ratio = tx_user_ratio.reset_index()
tx_user_ratio = tx_user_ratio.dropna()

#print the dafaframe
tx_user_ratio
```

```python
%matplotlib inline
```

```python
plt.figure(figsize=(10,6))
plt.title("New Customer Ratio")
plt.xlabel("Month")
plt.ylabel("Ratio")
sb.barplot(x= "Year_Qtr",y= "CustomerID", data=tx_user_ratio)
```
