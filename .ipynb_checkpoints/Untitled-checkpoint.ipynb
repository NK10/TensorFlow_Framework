```python
((train_data.isnull().sum()/len(train_data))*100).reset_index().sort_values(ascending = False,
                                                                         by = 0)
```

```python
col = ((train_data.isnull().sum()/len(train_data))*100).reset_index().sort_values(ascending = False,
                                                                         by = 0).iloc[0:4,0]
```

```python
train_data.drop(columns = col, inplace= True, axis = 1)
```

```python
train_data.dtypes.value_counts()
```

just to run it quickly,i used the mean and mode, but in actual we have to see the distribution of each categorical value and then fill it accordingly

```python
train_data.fillna(train_data.mean(), inplace=True)
```

the reason to do the iloc is because the mode returns the dataframe and mean return the series

```python
train_data.fillna(train_data.mode().iloc[0],inplace= True)
```
