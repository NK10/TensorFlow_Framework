Before classify every problem as machine learning and think it as a magic bullet. Always take a step back and understand if this require a machine learning or not. Basic rule of thumb is to see if the problem you are trying to solve has predefined steps and require a definite calculation and if that is the case then its better to code it explicitly rather using machine learning.

If the process has some degree of undertainity and require historical data to find patterns than it may be use case for machine learning.



High Level the machine learning the project is divided into 7 steps:

<b>  
1) understaning project requirements: This is one of the critical step, as in this we clarify what the business requirement is  
2) Data preparation  
3) Data understanding  
4) Model building  
5) Model evaluation  
6) Deployment of the model  
7) Monitoring  
 </b>
